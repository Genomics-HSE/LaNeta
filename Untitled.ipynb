```python
from ThLd import ThLd
import numpy as np
import matplotlib.pyplot as plt
from gen import Params
from theo import thld_6
from matplotlib import transforms

T1 = [0, 5, 10, 20, 40]
T2 = [2, 5, 10]


plt.subplots(figsize=(15.3, 25))

plt.subplots_adjust(hspace = None)

# first of all, the base transformation of the data points is needed
base = plt.gca().transData
rot = transforms.Affine2D().rotate_deg(90)

for i in range(len(T1)):
    for j in range(len(T2)):

        params = Params(T1[i], T2[j], 0.2, 0.2)

        name = 'saved_data/'+'T1-'+str(T1[i])+'_T2-'+str(T2[j])+'_M1-'+str(0.2)+'_M2-'+str(0.2)+'.npy'
        coef = np.load(name)
        P=300
        plt.subplot(len(T1), len(T2), i*3+j+1)
        #plt.subplots_adjust(hspace = 0.1, wspace=0.1)
        print(i*3+j+1)
        plt.imshow(np.rot90(coef[P//200:P//5,P//200:P//5]), cmap='hsv')
        plt.contour(np.rot90(thld_6(params)[P//200:P//5,P//200:P//5]), colors='black', linestyles = 'solid')
        if (j == 0):
            plt.ylabel('T1='+str(T1[i]), fontsize=35)
        if (i == len(T1) - 1):
            plt.xlabel('T2='+str(T2[j]), fontsize=35)
        plt.xticks([])
        plt.yticks([])

        
plt.tight_layout() # Or equivalently,  "plt.tight_layout()"

plt.savefig('img.jpeg')
```
