```python
import gen
import Algorithm
import numpy as np
from datetime import datetime

T1s = [40]
T2s = [2]

m1s = [0.2]
m2s = [0.2]

P = 300

startTime = datetime.now() 

coef = np.zeros((P, P))
for t1_i in range(len(T1s)):
    for t2_i in range(len(T2s)):
        for m1_i in range(len(m1s)):
            for m2_i in range(len(m2s)):
                data = gen.twopulse(T1s[t1_i], T2s[t2_i], m1s[m1_i], m2s[m2_i])
                coef = Algorithm.alg(data)
                
endTime = datetime.now() 
print(endTime - startTime)
```
